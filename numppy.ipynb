# *array manipulation*

```python
x1 = np.random.randint(10, size=6)
x2 = np.random.randint(10, size=(3, 4))
x3 = np.random.randint(10, size=(3, 4, 5)) # size, r, c
```

```python
print("x3 ndim:", x3.ndim)
print("x3 shape:", x3.shape)
print("x3 size:", x3.size)
print("x3 dtype:", x3.dtype)
print("x3 itemsize:",x3.itemsize, "bytes") # size of single element in array
print("x3 nbytes", x3.nbytes, "bytes") # overaall size of array
```

```python
x1, x1[:3], x1[3:] # [:x] - before, [x:] - after
```

```python
x1, x1[2:5]
```

```python
x1, x1[::2], x1[1::2] # step 2, step 2 starting index 1
```

```python
x1, x1[::-1], x1[5::-2] #  reversed, reversed every other from index 5
```
